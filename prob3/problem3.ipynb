```python
from IPython.display import display, Image, Markdown

# Displays the image with markdown caption
def show_example(img_path, caption):
    
    display(Image(filename=img_path))
    display(Markdown(caption))
```

```python
display(Markdown("# Problem 3: Dataset Types"))
display(Markdown("In this notebook, I will present examples of six different dataset types with visualizations."))
```

```python
show_example(
    "images/1d_time_series.png",  
    "**1D Time Series:** Example of the population in Ireland and Europe over time. The data is one-dimensional with the population evolving over time."
)
```

```python
show_example(
    "images/2d_scalar_field.png",
    "**2D Scalar Field:** Example of a heatmap of temperature across the graph. Each point in 2D space has one scalar value (temperature)."
)
```

```python
show_example(
    "images/3d_scalar_field.png",
    "**3D Scalar Field:** Example of 3D scalar field. It's similar to the temperature but this one is used to measure different altitudes, latitudes, and longitudes."
)
```

```python
show_example(
    "images/2d_vector_field.png",
    "**2D Vector Field:** Example of two-dimention vector field, which is used to display magnitude and direction."
)
```

```python
show_example(
    "images/3d_shape.png",
    "**3D Shape:** Example of a 3D bar chart. Data represents surfaces/volumes in three dimensions."
)
```

```python
show_example(
    "images/graph.png",
    "**Graph (not tree):** Example of a social network graph. Nodes represent people, edges represent relationships. " \
    "Graphs may has cycles unlike trees."
)
```
